```python
# Import libraries
import os
import re
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Global variables
ROPE_EMA_ALPHA = .2
ABS_EMA_ALPHA = 0.01
```

```python
def get_largest_step_file(directory, prefix):
    """
    directory: path to folder like Module6/4Layers/
    prefix: "losses_step_" or "tokens_step_"
    returns: (filepath, step_number) for the largest step
    """
    max_step = -1
    best_file = None

    for fname in os.listdir(directory):
        if fname.startswith(prefix) and fname.endswith(".npy"):
            m = step_regex.search(fname)
            if m:
                step = int(m.group(1))
                if step > max_step:
                    max_step = step
                    best_file = os.path.join(directory, fname)

    return best_file, max_step
```

```python
def plot_comparions(smooth_type=None, max_token_count=None):
    plt.figure(figsize=(10, 6))

    for p_idx, (parent, color) in enumerate(parents.items()):
        for l_idx, layer in enumerate(layers):
            d = os.path.join(parent, layer)

            # --- Find best loss file ---
            loss_file, loss_step = get_largest_step_file(d, "losses_step_")
            if loss_file is None:
                print(f"Warning: No loss files found in {d}")
                continue
            losses = np.load(loss_file)

            # --- Find best token-count file ---
            token_file, token_step = get_largest_step_file(d, "token_counts_step_")
            if token_file is None:
                print(f"Warning: No token files found in {d}")
                continue
            tokens = np.load(token_file)

            # Safety check: lengths must match
            if len(tokens) != len(losses):
                print(f"Warning: Length mismatch in {d}: "
                      f"tokens={len(tokens)}, losses={len(losses)}. "
                      "Will truncate to min length.")
                L = min(len(tokens), len(losses))
                tokens = tokens[:L]
                losses = losses[:L]
                
            if max_token_count is not None:
                tokens = tokens[tokens < max_token_count]
                losses = losses[:len(tokens)]
            
            if smooth_type == "ema":
                losses = ema_smooth(losses, ABS_EMA_ALPHA if p_idx == 0 and l_idx != 1 else ROPE_EMA_ALPHA)

            label = f"{parent} â€“ {layer}"
            plt.plot(tokens, losses,
                     alpha=layer_alphas[layer],
                     linestyle=layer_styles[layer],
                     color=color,
                     label=label)


    plt.xlabel("Token Count")
    plt.ylabel("Loss")
    prefix = ''
    if smooth_type == 'ema':
        prefix = 'Exponential Moving Average Smoothed '
    if smooth_type == 'moving':
        prefix = 'Moving Window Average Smoothed '
    plt.title(f"{prefix}Loss vs Token Count: Module6 vs FinalProject Across Model Depths")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
# Parent dirs: color per parent
parents = {
    "..\BaseModel": "C0",
    "..\FinalProject": "C3",
}

# Subdirectories: line style per depth
layers = ["2Layers", "4Layers", "8Layers", "12Layers"]
layer_styles = {
    "2Layers": "-",
    "4Layers": "--",
    "8Layers": "-.",
    "12Layers": ":"
}

layer_alphas = {
    "2Layers": 0.2,
    "4Layers": 0.4,
    "8Layers": 0.6,
    "12Layers": 1.0
}

# Regex to extract step number from filenames like "losses_step_2000.npy"
step_regex = re.compile(r"_(\d+)\.npy$")

plot_comparions()
```

```python
def ema_smooth(values, alpha=0.03):
    smoothed = []
    prev = values[0]
    for v in values:
        prev = alpha * v + (1 - alpha) * prev
        smoothed.append(prev)
    return np.array(smoothed)
```

```python
plot_comparions(smooth_type='ema')
```

```python
plot_comparions(smooth_type='ema', max_token_count=80000000)
```
